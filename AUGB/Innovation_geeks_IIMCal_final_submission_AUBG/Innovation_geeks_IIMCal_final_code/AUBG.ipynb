```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.decomposition import PCA
```

```python
df1 = pd.read_csv("D:/Suman/PGDBA/Competitions/AUBG/data.csv")
```

```python
df2 = pd.read_csv("D:/Suman/PGDBA/Competitions/AUBG/winner.csv")
```

```python
data = pd.read_csv("D:/Suman/PGDBA/Competitions/AUBG/5dc975b0250f1_Training_Dataset.csv")
```

```python
pca = PCA(n_components=5)
pca.fit(df1)
```

```python
print(pca.explained_variance_ratio_) 
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df1)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, df2['AUBG_Dataset.Winner']], axis = 1)
```

```python
finalDf.head()
```

```python
fig = plt.figure(figsize=(8,5))
plt.scatter(finalDf['principal component 1'],finalDf['principal component 2'],c=finalDf['AUBG_Dataset.Winner'],cmap = 'viridis')
fig.savefig("PCA_Analysis")
```

```python
finalDf1 = finalDf[finalDf['AUBG_Dataset.Winner']==0]
finalDf2 = finalDf[finalDf['AUBG_Dataset.Winner']==2]
finalDf3 = finalDf[finalDf['AUBG_Dataset.Winner']==1]
```

```python
fig = plt.figure(figsize=(8,5))
plt.scatter(finalDf1['principal component 1'],finalDf1['principal component 2'],c=finalDf1['AUBG_Dataset.Winner'],cmap = 'viridis')
fig.savefig("PCA_Analysis_1")
```

```python
fig = plt.figure(figsize=(8,5))
plt.scatter(finalDf2['principal component 1'],finalDf2['principal component 2'],c=finalDf2['AUBG_Dataset.Winner'],cmap = 'viridis')
fig.savefig("PCA_Analysis_2")
```

```python
fig = plt.figure(figsize=(8,5))
plt.scatter(finalDf3['principal component 1'],finalDf3['principal component 2'],c=finalDf3['AUBG_Dataset.Winner'],cmap = 'viridis')
fig.savefig("PCA_Analysis_3")
```
